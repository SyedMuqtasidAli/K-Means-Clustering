Report:
I have started with K values of 2 in which Image was divided into 2 clusters in which every color was according to the mean value of cluster 1 or cluster 2 and the final image was a result of only 2 colors image.

When we used k=3 and dominance of colors increased which makes image looks better as now there was improved image as now the image was divided into 3 clusters and 3 colors were shown.

For k-5, 5 clusters were made the image used 5 colors and representation of the image was much improved from 2/3 clusters

Similarly fo 10,15,20 clusters we had increased the image colors and representation of the image also improved a lot.

```python
import numpy as np
from sklearn.cluster import KMeans
from PIL import Image
```

```python
# Load the image
image = Image.open("face.jpg")
# Convert the image to a numpy array
image_array = np.array(image)
image_array
```

```python
# Reshape the image array to 2D
height, width, channels = image_array.shape
image_2d = image_array.reshape(height * width, channels)
image_2d
```

```python
# Define the value of K =2,3,5,10,20
k = 20
```

```python
# Create a KMeans instance
kmeans = KMeans(n_clusters=k, random_state=0)
```

```python
# Fit the K-means model to the data
kmeans.fit(image_2d)
```

```python
# Get the labels for each pixel
labels = kmeans.labels_
labels
```

```python
# Get the cluster centers (colors)
colors = kmeans.cluster_centers_
colors
```

```python
# Replace each pixel with its corresponding cluster center color
compressed_image_2d = colors[labels]
compressed_image_2d
```

```python
# Reshape the compressed image array to its original shape
compressed_image_array = compressed_image_2d.reshape(height, width, channels)
compressed_image_array
```

```python
# Convert the compressed image array back to Image format
compressed_image = Image.fromarray(compressed_image_array.astype(np.uint8))
compressed_image
```

```python
# Save the compressed image
compressed_image.save("compressed_image_k20.jpg")
```

```python
import matplotlib.pyplot as plt
# List of image paths
image_paths = ["compressed_image_k2.jpg",
               "compressed_image_k3.jpg",
               "compressed_image_k5.jpg",
               "compressed_image_k10.jpg",
               "compressed_image_k15.jpg",
               "compressed_image_k20.jpg",
               ]

# Iterate over the image paths
for path in image_paths:
    # Load the image
    image = Image.open(path)

    # Display the image
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
